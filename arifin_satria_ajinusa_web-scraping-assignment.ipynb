#Scrapping using Beautifulsoup

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

# antara news

```python
url = 'https://www.antaranews.com/top-news'

response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
# soup
```

```python
# list_content = soup.select('.m-table-kurs') #bca kurs
# list_content = soup.select('body') #bca kurs
list_content = soup.select('.wrapper__list__article') # antara
# list_content = soup.select('article')
```

```python
len(list_content)
```

```python
list_content[0]
# list_content
```

```python
results = []
for item in list_content:
  data = dict()

  # data['curr'] = item.select_one('.tb-cell').text
  # data['buy'] = item.find(attrs={"rate-type": "ERate-buy"}).text

  data['title'] = item.select_one('.card__post__title').text
  data['time'] = item.select_one('.card__post__author-info').text
  data['url'] = item.select_one('.post_title a').get('href')
  results.append(data)

pd.DataFrame(results)
```

# BCA - kurs IDR to xxx

```python
url = 'https://www.bca.co.id/id/informasi/kurs'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
# soup
```

```python
list_content = soup.select('.m-table-kurs') #bca kurs
# list_content = soup.select('body') #bca kurs
# list_content = soup.select('.wrapper__list__article') # antara
# list_content = soup.select('article')
```

```python
len(list_content)
```

```python
list_content[0]
# list_content
```

```python
table = soup.find('table', class_ = 'm-table-kurs')
# print(table)

results = []
for employee_data in table.find_all('tbody'):
  rows = employee_data.find_all('tr')
  # print(rows)

  for row in rows:
    # name = row.find_all('td')[0].text
    # print(name)

    data = dict()
    data['curr'] = row.find_all('td')[0].text
    data['buy'] = row.find_all('td')[1].text
    data['sell'] = row.find_all('td')[2].text

    results.append(data)

pd.DataFrame(results)
```
